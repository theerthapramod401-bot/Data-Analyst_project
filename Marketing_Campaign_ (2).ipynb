**11 . Remove Duplicates**

```python
df=df.dropna()
```

**12 . Convert Date Column**

```python
df['Date'] = pd.to_datetime(df['Date'])
```

**13 . Create CTR Feature**

CTR = Clicks ÷ Impressions × 100

```python
df['CTR'] = (df['Clicks'] / df['Impressions']) * 100
```

```python
df['Clicks'].sum()
```

**15 . Total Impressions**

**18 . Statistical Summary**

```python
df.describe()
```

**19 . Channel-wise Performance**

```python
channel_summary = df.groupby('Channel_Used').agg({
    'Clicks':'sum',
    'Impressions':'sum',
    'ROI':'mean',
    'Acquisition_Cost':'sum'
})

channel_summary
```

**20 . Campaign Type Analysis**

```python
campaign_summary = df.groupby('Campaign_Type').agg({
    'ROI':'mean',
    'Clicks':'sum'
})

campaign_summary
```

**21 . Campaign Type Distribution**

```python
df['Campaign_Type'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title('Campaign Type Distribution')
plt.show()
```

**22 . ROI vs Acquisition Cost**

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Acquisition_Cost'], df['ROI'])
plt.title('ROI vs Acquisition Cost')
plt.xlabel('Acquisition Cost')
plt.ylabel('ROI')
plt.grid(True)
plt.show()
```

* High acquisition cost does not necessarily lead to high ROI.
* Several campaigns achieve strong ROI with relatively lower spending, indicating efficient resource utilization.
* Some high-cost campaigns generate only moderate returns, suggesting inefficiencies.

**23 . Correlation Heatmap**

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(
    df[['Clicks','Impressions','ROI','Acquisition_Cost','CTR']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title('Correlation Heatmap')
plt.show()
```

* Clicks and Impressions typically show a strong positive correlation.
* CTR may positively influence ROI, indicating that higher engagement often leads to better performance.
* Acquisition Cost may have a weak relationship with ROI, suggesting that spending more does not automatically improve returns.

24 . ROI by Campaign Type

```python
campaign_roi = df.groupby('Campaign_Type')['ROI'].mean()

plt.figure(figsize=(8,5))
campaign_roi.plot(kind='bar')
plt.title('Average ROI by Campaign Type')
plt.xlabel('Campaign Type')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.show()
```

* Some campaign types consistently outperform others in terms of average ROI.
* Lower-performing campaign types consume resources without generating comparable returns

25 . Monthly Campaign Trend

```python
df['Month'] = df['Date'].dt.to_period('M')

monthly_clicks = df.groupby('Month')['Clicks'].sum()

monthly_clicks.plot(figsize=(10,5))
plt.title('Monthly Click Trend')
plt.xlabel('Month')
plt.ylabel('Total Clicks')
plt.show()
```

* Certain months show stronger ROI, indicating periods of higher customer responsiveness.
* Some months experience performance declines, possibly due to seasonality or market conditions.

**26 . Top 10 Campaigns by ROI**

```python
top_campaigns = df.nlargest(10, 'ROI')

plt.figure(figsize=(10,5))
plt.bar(top_campaigns['Campaign_ID'].astype(str),
        top_campaigns['ROI'])
plt.title('Top 10 Campaigns by ROI')
plt.xlabel('Campaign ID')
plt.ylabel('ROI')
plt.xticks(rotation=45)
plt.show()
```

* The top campaigns achieved substantially higher ROI than the average campaign.
* These campaigns likely benefited from effective targeting, channel selection, and budget allocation.
